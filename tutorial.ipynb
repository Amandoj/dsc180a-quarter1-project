```python
import qiime2, biom, pandas as pd
pd.options.mode.chained_assignment = None
from qiime2.plugins.sample_classifier.pipelines import classify_samples
```

```python
fungi_decon = pd.read_csv('data/count_data_fungi_decontaminated_HMS_1986samples.tsv', sep='\t', header=0)
fungi_decon.head()
```

```python
fungi_high = pd.read_csv("data/count_data_fungi_high_coverage_HMS_1986samples.tsv",sep='\t',header= 0)
fungi_high.head()
```

```python
fungi_intersect = pd.read_csv("data/count_data_fungi_WIS_intersect_HMS_1986samples.tsv",sep="\t",header=0)
fungi_intersect.head()
```

```python
fungi_metadata = pd.read_csv("data/metadata_fungi_HMS_1986samples.tsv",sep="\t",header=0)
fungi_metadata.head()
```

```python
taxonomy = pd.read_csv("data/taxonomy_table_rep200.tsv",sep="\t",header=0)
taxonomy.head()
```

```python

metadata_ml = fungi_metadata[['sampleid','disease_type','sample_type']]
metadata_ml_tumor = metadata_ml[metadata_ml['sample_type']=='Primary Tumor']
metadata_ml_tumor.head()
```

```python
disease_type_counts = metadata_ml_tumor['disease_type'].value_counts()
disease_type_counts.index
```

```python
def organize_data_for_ml(dataset):
    for dt in disease_type_counts.index:
        temp_meta = metadata_ml_tumor
        temp_meta['predY'] = temp_meta['disease_type'].apply(lambda disease: disease if disease == dt else "OtherCancerType")
        predY_counts = temp_meta['predY'].value_counts() 
        if any(predY_counts < 20): continue
        
        minorityClassSize = predY_counts.min()
        majorityClassSize = predY_counts.max()
        
        minorityClassName = predY_counts.idxmin()
        majorityClassName = predY_counts.idxmax()
        
        trainX = pd.merge(left=dataset,right=temp_meta['sampleid'], on='sampleid',how='inner')
        trainY = temp_meta.set_index('sampleid')[['predY']]
        # convert pandas dataframes to Artifact and Metadata in order to use with qiime2
    return qiime2.Artifact.import_data("FeatureTable[Frequency]", trainX.set_index('sampleid')), qiime2.Metadata(trainY)
```

```python
trainX,trainY = organize_data_for_ml(fungi_high)
```

```python
results = classify_samples(table=trainX,metadata=trainY.get_column('predY'),cv = 10, 
                           estimator='GradientBoostingClassifier', n_estimators = 150,
                           parameter_tuning=False, random_state=42, test_size=.1)
```
